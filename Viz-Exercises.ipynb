```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# From the Python graph gallery:

df = sns.load_dataset('iris')
_ = sns.boxplot(x='species', y='sepal_length', data=df)
_ = sns.swarmplot(x='species', y='sepal_length', data=df, color="grey")
```

* Run the cell below. Which plotting libraries are used? Comment on every line. 
* Creating quick and dirty plots like these can be powerful EDA. Using data from this lab, use the Pandas `.plot`  method to visualize a single variable. Then, use the same method to visualize the results of a Pandas `.groupby` operation.

```python
df = sns.load_dataset('tips')
_ = df.groupby('sex')['tip'].mean().plot(kind='bar')
_ = plt.title('Average Tip By Customer Sex')
```
